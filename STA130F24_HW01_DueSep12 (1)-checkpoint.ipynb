Definition of Attributes: Properties or qualities of an object that store information are called attributes. Parentheses are not used; they are accessed directly.
Application: Features give details about the object or its condition. Instead of doing anything, they just return the data that is stored inside the object.
Take df.shape as an example.
One of the properties of the DataFrame df is df.shape. It gives back a tuple with the DataFrame's dimensions (row and column counts).
It can be accessed without brackets since it is an attribute: df.shape.
Definition of Methods: A method is a function that is connected to an object. They do out operations or calculations, change the item, or provide fresh information. Parentheses are used to call methods.Usage: Methods compute values based on the data in the object or carry out operations on the object. They are also capable of changing their behaviour through arguments.
Example: The DataFrame df has a function called df.describe(). It calculates and provides the DataFrame's summary statistics, such as the mean, standard deviation, and count for the numeric columns.
It is called with parenthesis since it is a method: df.describe().
Key distinctions between action and access:

Feature: Offers quick access to data that has been stored. You omit the brackets.
As an illustration, the DataFrame's shape is provided via df.shape.
Method: carries out an operation or a function. The procedure is called using brackets.
For instance, the DataFrame's descriptive statistics are produced by df.describe().
Usability:

Attribute: Provides the object's stored data in a simple manner.
For instance, df.shape yields the number of rows and columns.

Approach: carries out a calculation or transformation and could provide a fresh value or item in response.
For instance, df.describe() generates a new DataFrame with summary statistics after doing statistical computations.
Grammar:

Accessed without the use of brackets.
Column names are returned by df.columns, for instance.
Method: Parenthetically called.
For instance, the first few rows of the DataFrame are returned by df.head().

 The counts you get from df.describe() and df.groupby("col1")["col2"].describe() reflect different aspects of the dataset due to how they handle missing values and grouping.

Understanding df.describe()
df.describe():

Count: This provides the number of non-null entries for each column.
Functionality: It gives summary statistics for numeric columns by default, including:
Count: The number of non-null values.
Mean: Average of the values.
Std: Standard deviation.
Min/Max: Minimum and maximum values.
25%/50%/75%: Percentiles (25th, 50th, 75th).
Effect of Missing Data:

Missing Data: If a column has missing values, the count in df.describe() will be less than the total number of rows in the DataFrame. For example, if a column has 100 rows but 10 missing values, df.describe() will show a count of 90 for that column.
Understanding df.groupby("col1")["col2"].describe()
df.groupby("col1")["col2"].describe():

GroupBy Operation: This groups the DataFrame by unique values in col1 and then performs descriptive statistics on col2 for each group.
Count: The count here is the number of non-null entries for col2 within each group.
Effect of Missing Data:

Per Group: The count value in this context shows the number of non-null values within each group formed by col1. If some groups have missing values in col2, each group's count will reflect the number of non-null values within that specific group only.
Key Differences
Scope of Count:

df.describe():

Global Count: Provides a single count of non-null values across the entire column in the DataFrame, not considering any grouping.
Effect: A column with missing values will show a reduced count compared to the total number of rows.
df.groupby("col1")["col2"].describe():

Grouped Count: Provides counts of non-null values within each group defined by col1. Each group will have its own count based on the subset of data within that group.
Effect: Counts will vary from group to group depending on how many non-null values are present in each specific group.
Handling Missing Data:

df.describe():

Aggregates the count of non-null entries across the whole DataFrame, not considering how data is distributed across different groups.
df.groupby("col1")["col2"].describe():

Handles missing data on a per-group basis, showing how missing values are distributed among different groups.
